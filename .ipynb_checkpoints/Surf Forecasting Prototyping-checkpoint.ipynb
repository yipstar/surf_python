```python
OBS_ENDPOINT = "https://sdf.ndbc.noaa.gov/sos/server.php"
#station_id = "46221"
station_id = "42040"

eventtime = "latest"
observation = 'Waves'
```

```python
params = {
            'offering': 'urn:ioos:station:wmo:{}'.format(station_id),
            'observedproperty': observation,
            'eventtime': eventtime
        }
```

```python
default_params = {
        'request': 'GetObservation',
        'service': 'SOS',
        'version': '1.0.0',
        'responseformat': 'text/csv',
    }
```

```python
import requests
params.update(default_params)
response = requests.get(OBS_ENDPOINT, params=params)
```

```python
print(response.text)
```

```python
import csv
from io import StringIO
reader = csv.DictReader(StringIO(response.text))
```

```python
row = next(reader)
```

```python
print(row['station_id'])
#print(row['spectral_energy (m**2/Hz)'])
#print(row)
```

```python
spectral_energy = 'spectral_energy (m**2/Hz)'
print(row[spectral_energy])
```
